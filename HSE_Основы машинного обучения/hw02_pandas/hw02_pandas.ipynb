**Задание 10. Ответьте на вопросы:**
    * Какая доля студентов сдала экзамен по математике (passmark >= 50)?
    * Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?
    * Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике? 

```python
#Какая доля студентов сдала экзамен по математике (passmark >= 50)?

a1 = float((df[['math score']] >= passmark).sum()/len(df[['math score']]))*100

print(f'{a1:0.2f}% студентов сдали экзамен по математике')
```

```python
#Какая доля студентов, проходивших курс подготовки к экзамену, сдала экзамен по математике?

a2 = float(((df[['test preparation','math score']].query(
    '`math score` >= @passmark').groupby(
        'test preparation').count()).loc['completed']/
            (df[['test preparation','math score']].groupby(
                'test preparation').count())).loc['completed'])*100

print(f'{a2:0.2f}% студентов, проходивших курс подготовки к экзамену, сдали экзамен по математике')
```

```python
#Какая доля женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике?

a3 = float(((df[df.gender=='female'][['test preparation','math score']].query(
    '`math score` < @passmark').groupby(
        'test preparation').count()).loc['none'])/
           (df[df.gender=='female'][['test preparation','math score']].groupby(
               'test preparation').count()).loc['none'])*100

print(f'{a3:0.2f}% женщин, не проходивших курс подготовки к экзамену, не сдала экзамен по математике')
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
def GetGrade(average_mark):
    avr_m = np.where(average_mark < 50, 'F',
                    np.where(average_mark < 60, 'E',
                            np.where(average_mark < 70, 'D',
                                    np.where(average_mark < 80, 'C',
                                            np.where(average_mark < 90, 'B',
                                                    np.where(average_mark >= 90, 'A', 'NaN'))))))
    return avr_m

# your code here
```

```python
df['Grade'] = GetGrade(df.average_mark)

df.head(10)
```
